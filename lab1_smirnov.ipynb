```python
import pandas
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
dataframe = pandas.read_csv("iris.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]

X
```

```python
Y
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
dummy_y = to_categorical(encoded_Y)
# encoded_Y
dummy_y
```

```python
epochs_arr = [x for x in range(1, 76)]
```

```python
model_1 = Sequential()
model_1.add(Dense(4, activation='relu'))
model_1.add(Dense(3, activation='softmax'))
```

```python
model_1.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
fit_arr_1 = model_1.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1)
train_loss_1 = fit_arr_1.history['loss']
train_acc_1 = fit_arr_1.history['accuracy']
```

```python
plt.figure()
plt.plot(epochs_arr, train_loss_1)
plt.suptitle("Loss model 1")
```

```python
plt.figure()
plt.plot(epochs_arr, train_acc_1)
plt.suptitle("Accuracy model 1")
```

```python
model_1.predict(X)
```

```python
model_2 = Sequential()
model_2.add(Dense(3, activation='relu'))
model_2.add(Dense(3, activation='softmax'))
model_2.add(Dense(3, activation='sigmoid'))
# model_2.add(Dense(3, activation='selu'))
# model_2.add(Dense(3, activation='linear'))
```

```python
model_2.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
fit_arr_2 = model_2.fit(X, dummy_y, epochs=len(epochs_arr), batch_size=10, validation_split=0.1)
train_loss_2 = fit_arr_2.history['loss']
train_acc_2 = fit_arr_2.history['accuracy']
```
